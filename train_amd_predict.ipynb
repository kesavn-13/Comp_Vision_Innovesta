```python
from PIL import Image
from torchvision.transforms import ToTensor
from rembg import remove
import torchvision
import torch
from ultralytics import YOLO
import matplotlib.pyplot as plt
import os

# Load the model
model = YOLO(r'E:\comp_vision\YOLOLO\runs\classify\train9\weights\last.pt')  # load a custom model

# Folder containing test images
test_folder = r"C:\Users\167093\Downloads\new_test"

# Iterate through each file in the folder
for filename in os.listdir(test_folder):
    if filename.endswith(".png") or filename.endswith(".jpg"):
        # Load the image
        image_path = os.path.join(test_folder, filename)
        image = Image.open(image_path)

        # Perform background removal
        output = remove(image)

        # Remove the alpha channel if present
        if output.mode == 'RGBA':
            image1 = output.convert('RGB')
        else:
            image1 = output

        # Resize the image
        resized_img = image1.resize((640, 640))

        # Make a prediction
        results = model(resized_img)

        # Display or save the results for each image
        plt.imshow(resized_img)
        plt.title(f"Results for {filename}")
        plt.show()
```
