```python
import os
import cv2
import numpy as np
import random

# Set the path to the directory containing the images
directory_path = path

# Get a list of all files in the directory
all_files = os.listdir(directory_path)

# Filter only image files (you may adjust the file extensions as needed)
image_files = [file for file in all_files if file.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Choose 20 random images
random_images = random.sample(image_files, 50)

# Create the output directory if it doesn't exist
output_directory = os.path.join(directory_path, "noisy_images")
os.makedirs(output_directory, exist_ok=True)

# Iterate over the randomly selected images
for idx, image_file in enumerate(random_images):
    # Read the image
    image_path = os.path.join(directory_path, image_file)
    img = cv2.imread(image_path)

    # Generate random noise and add it to the image
    noise = np.random.normal(0, 50, img.shape)
    noisy_img = np.clip(img + noise, 0, 255).astype(np.uint8)

    # Save the noisy image
    output_path = os.path.join(output_directory, f"noise{idx + 1}.jpg")
    cv2.imwrite(output_path, noisy_img)

print("Adding noise process completed.")
```

```python
import os
import cv2
import numpy as np
import random

# Set the path to the directory containing the images
directory_path = path

# Get a list of all files in the directory
all_files = os.listdir(directory_path)

# Filter only image files (you may adjust the file extensions as needed)
image_files = [file for file in all_files if file.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Choose 50 random images
random_images = random.sample(image_files, 50)

# Create the output directory if it doesn't exist
output_directory = os.path.join(directory_path, "contrast_brightness_images")
os.makedirs(output_directory, exist_ok=True)

# Iterate over the randomly selected images
for idx, image_file in enumerate(random_images):
    # Read the image
    image_path = os.path.join(directory_path, image_file)
    img = cv2.imread(image_path)

    # Generate random contrast and brightness values
    contrast = random.uniform(0.5, 1.5)  # You can adjust the range as needed
    brightness = random.randint(-50, 50)  # You can adjust the range as needed

    # Apply contrast and brightness adjustment
    adjusted_img = np.clip(contrast * img + brightness, 0, 255).astype(np.uint8)

    # Save the adjusted image
    output_path = os.path.join(output_directory, f"cn{idx + 1}.jpg")
    cv2.imwrite(output_path, adjusted_img)

print("Contrast and brightness adjustment process completed.")
```

```python
import os
import cv2
import numpy as np
import random

# Set the path to the directory containing the images
directory_path = path

# Get a list of all files in the directory
all_files = os.listdir(directory_path)

# Filter only image files (you may adjust the file extensions as needed)
image_files = [file for file in all_files if file.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Choose 50 random images
random_images = random.sample(image_files, 50)

# Create the output directory if it doesn't exist
output_directory = os.path.join(directory_path, "hue_saturation_images")
os.makedirs(output_directory, exist_ok=True)

# Iterate over the randomly selected images
for idx, image_file in enumerate(random_images):
    # Read the image
    image_path = os.path.join(directory_path, image_file)
    img = cv2.imread(image_path)

    # Generate random hue and saturation values
    hue_shift = random.randint(-20, 20)  # You can adjust the range as needed
    saturation_scale = random.uniform(1, 10)  # You can adjust the range as needed

    # Convert the image to HSV color space
    img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    # Apply hue shift and saturation scaling
    img_hsv[:, :, 0] = (img_hsv[:, :, 0] + hue_shift) % 180
    img_hsv[:, :, 1] = np.clip(img_hsv[:, :, 1] * saturation_scale, 0, 255)

    # Convert the image back to BGR color space
    adjusted_img = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR)

    # Save the adjusted image
    output_path = os.path.join(output_directory, f"hs{idx + 1}.jpg")
    cv2.imwrite(output_path, adjusted_img)

print("Hue and saturation adjustment process completed.")
```
